#Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# python script to load csv and read it as a dataframe

```python
autodf=pd.read_csv('autos.csv',encoding='utf-8')
```

#finding average price of autos using mean()  

```python
avg_price= autodf['price'].mean()
print('Average price of autos:',avg_price)
```

#listing our the different types of vehicles using unique()

```python
vehicle_type=autodf['vehicleType'].unique()
print('Different possible types of vehicles:',vehicle_type)
```

#Find and print standard deviation of column kilometer using std()


```python
std_km = autodf['kilometer'].std()
print('Standard deviation of kilometer column:', std_km)
```

#Draw a bar graph to represent count of different type of column brand


```python
brand_cnts=autodf['brand'].value_counts()
plt.bar(brand_cnts.index,brand_cnts)
plt.xticks(rotation=90)
plt.xlabel('Brand')
plt.ylabel('Count')
plt.title('Count of different Types of Brands')
plt.show()
```

#Find out which VehicleType  is sold minimum and maximum using value_counts()

```python
vehicle_cnts = autodf['vehicleType'].value_counts()

min_cnt = vehicle_cnts.min()
min_vehicle_type = vehicle_cnts[vehicle_cnts == min_cnt].index[0]
# retruns boolean series indicating the elements of vehicle_cnt are equal to mac_cnt and then have the max count
max_cnt = vehicle_cnts.max()
max_vehicle_type = vehicle_cnts[vehicle_cnts == max_cnt].index[0]

print('Vehicle type with minimum sales:', min_vehicle_type)
print('Vehicle type with maximum sales:', max_vehicle_type)
```


#Create a pie chart to represent different types of  gearbox count 

```python
gearbox_cnts=autodf['gearbox'].value_counts()
plt.pie(gearbox_cnts,labels=gearbox_cnts.index,radius=1,autopct='%1.1f%%')
plt.title("Different greabox counts")
```

# Calculate and print lowest yearOfRegistration  using min() and highest yearOfRegistration using max()

```python

lowest_year = autodf['yearOfRegistration'].min()
highest_year = autodf['yearOfRegistration'].max()
print('Lowest year of registration:', lowest_year)
print('Highest year of registration:', highest_year)
```
